```python
#Q1 Load the Sample 1 CSV File

import pandas as pd
sp1 = pd.read_csv(r"C:\Users\ram singh\Desktop\Term2_Python_Project\sample1_group24_035043_035037.csv", low_memory=False)
sp1
```

```python
for column_headers in sp1.columns: 
    print(column_headers)
```

```python
#Q2 Generate a Report on Feature-wise Missing Data in the Sample

missing_values = sp1.isnull()

# Use the sum() function to sum up the number of missing values in each column
feature_wise_missing_data = missing_values.sum()

# Print the report
print(feature_wise_missing_data)
```

```python
#Q3 Treat Missing Data in the Sample
#Q4 Recode Features with Text Data into Numeric Data in the Sample (Do Not Delete the Features with Text Data. Create New Features with Equivalent Numeric Data). Use the Recoded Features for all Analysis

recodes = {"False": 0, "True": 1}
recodes_type ={"A": 1, "B": 2, "C": 3}
sp1 = sp1.assign(new_colum_IsHoliday = sp1['IsHoliday'])
sp1 = sp1.assign(new_column_Type = sp1['Type'])
sp1["new_colum_IsHoliday"]=sp1["new_colum_IsHoliday"].apply(str)
sp1["new_column_Type"]=sp1["new_column_Type"].apply(str)

sp1["new_colum_IsHoliday"]=sp1["new_colum_IsHoliday"].map(recodes)
sp1["new_column_Type"]=sp1["new_column_Type"].map(recodes_type)

print(sp1["new_colum_IsHoliday"])
print(sp1["new_column_Type"])
sp1 = sp1.fillna(sp1.mean())
#sp1["Date"] = sp1["Date"].fillna('05/02/2010')
print(sp1)
```

```python
#Q5 Save the Processed Sample 1 as Sample 2 (sample1_xxx_yyy.csv) in form of CSV File in the Local Folder Location

sp1.to_csv("sample2_group24_035043_035037.csv")
```
