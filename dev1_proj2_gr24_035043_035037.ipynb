```python
# Generate a Comprehensive Report on the Insights Drawn from the Sample about Store Performance (and/or any other Insight)

import pandas as pd
import matplotlib.pyplot as plt

sp2.describe()
```

```python
import seaborn as sns 

g = sns.catplot(data=sp2, x='Store', y='Weekly_Sales', color ='green')
g.fig.set_figwidth(15)
g.fig.set_figheight(8)
```

```python
# Total Weekly_Sales of Store

Weekly_Sales = sp2.groupby('Store')['Weekly_Sales'].sum()
Weekly_Sales
```

```python
Weekly_Sales.plot(kind='bar')
```

```python
# Total Dept of Store

Dept = sp2.groupby('Store')['Dept'].sum()
Dept
```

```python
Dept.plot(kind='bar')
```
